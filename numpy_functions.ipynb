important functions 

zeroes functions

```python
import numpy as np
arr=np.zeros((2,2,10))
print(arr)
# ones function
arr=np.ones((1,2,3)) # layers,rows,cols
print(arr)
```

2) full functions

```python
arr=np.full((2,3,4),9)
print(arr)
# identity function
arr=np.eye(3)
print(arr)
# empty function
arr=np.empty((2,3))
print(arr)
```

3) arrange function
<br>arange(start , stop,step)

```python
arr=np.arange(0,101,2.5) # arange(start , stop ,step)
print(arr)
arr=np.linspace(0,10,20) # start,stop,step
print(arr)
```
