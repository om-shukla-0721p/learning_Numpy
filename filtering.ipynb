1) filtering= refers to the process of selecting elements from an array that match a given condition

```python
import numpy as np
# 1) boolean indexing -> shape not preserved
ages=np.array([[34,12,45,17,16,65,14,29],
      [19,42,52,98,54,21,10,20]])

teenagers=ages[ages<18] # boolean indexing -> retunred flattened 1D array
adults=ages[(ages>=18) & (ages<65)]

elders=ages[ages>=65]
non_adults=ages[(ages<18) | (ages>=65)]
even_ages=ages[ages%2==0]
odd_ages=ages[ages%2!=0]
print(teenagers)
print(adults)
print(elders)
print(non_adults)
print(even_ages)
print(odd_ages)
```

2) where function to preserve the shape of a function
 
np.where(condition,value_if_true,value_if_false)

is a lot slower than boolean indexing though 

only use if shape needs to be reserved

```python
ages=np.array([[34,12,45,17,16,65,14,29],
      [19,42,52,98,54,21,10,20]])

ages[ages<=18]=0 # updates the original array
adults=np.where((ages>=18)& (ages<65),ages,-1) # creates a copy of origna arr with 
                                               # given updates
print(ages)
print(adults)
```
