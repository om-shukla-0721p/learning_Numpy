1) random used for generating random numbers 

used for test cases and others


random(start,end,size)

```python
import numpy as np
# rng =Random Number Generater
rng=np.random.default_rng(seed=1)
arr=rng.integers(low=1,high=101,size=(4,4))
print(arr)
```

2) uniform


```python
np.random.seed(seed=1)
print(np.random.uniform(low=-1,high=1,size=(3)))
```

3) shuffling arrays

```python
rng=np.random.default_rng()# seed=1
arr=[1,2,3,4,5]
rng.shuffle(arr)
print(arr)
```

4) choice


```python
fruits=np.array(["apple","banana","pineapple","orange","strawberry"])
rng=np.random.default_rng()
fruit=rng.choice(fruits,size=(3,2))
print(fruit)
```
