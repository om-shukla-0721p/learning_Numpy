1) scalar arithmatic


```python
# scalar arithmatic
# is used to do basic operations to an array
import numpy as np

arr=np.array([1,2,3])

print(arr+1)
print(arr-2)
print(arr*2)
print(arr/2)
print(arr**5)
```

2. vectorized math functions

```python
# vectorized math func
arr2=np.array([1.01,3.43,2891.21])
print(np.sqrt(arr2))
print(np.floor(arr2))
print(np.ceil(arr2))
print(np.pi)
```

```python
# exercise to print the area given the radii
radii=np.array([1,2,3])

print(np.square(radii)*np.pi)
```

3) element wise arithmatic


```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])

print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1**arr2)
print(arr2**arr1)
```

4) comparision operaters


```python
# comparision operaters
scores=np.array([91,22,60,55,89,100])

print(scores==100)
print(scores>60)
print(scores<=60)
scores[scores<=60]=0
print(scores)
```
