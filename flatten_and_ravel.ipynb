1) reverse an array

```python
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
print(np.flip(arr,axis=0))
print(np.flip(arr,axis=1))
```

specific row or col reverse

```python
arr[2]=np.flip(arr[2])
print(arr)
arr[:,1]=np.flip(arr[:,1])
print(arr)
```

flattening an array

```python
ones=np.ones((2,2,2),dtype=int)
print(ones)
print(ones.flatten())
```

ravel method 

changes are in place


```python
ones=np.ones((2,2,2),dtype=int)
# print(ones)
new_ones=ones.ravel()
new_ones[0]=999
print(ones)
print(new_ones)
```
